### Import function

```python
from dsmlibrary.datanode import DataNode
import json
from tqdm.auto import tqdm
import pandas as pd
```

### ใส่ค่า rawData , processData และเรียกใช้ datanode

```python
dir_raw_id = 71
dir_process_id = 73
```

```python
datanode = DataNode()
```

### แสดงที่อยู่ข้อมูลและเรียกดูผลข้อมูลที่ได้จากการ Scrapy

```python
meta, fp = datanode.get_file(datanode.get_file_id(directory_id=dir_raw_id, name='HERB_KASIKORNRESEARCH.json'))
meta
```

```python
data = json.load(fp)
data
```

### แสดงผลข้อมูลในรูปแบบตาราง

```python
df = pd.json_normalize(data)
df.head()
```

### Upload ลง Dataplatform

```python
datanode.write(df=df, directory=dir_process_id, name="HERB_KASIKORNRESEARCH", profiling=True, lineage=[datanode.get_file_id(directory_id=dir_raw_id, name='HERB_KASIKORNRESEARCH.json')])
```

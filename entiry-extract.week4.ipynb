### Import function

```python
from dsmlibrary.datanode import DataNode
import json
from tqdm.auto import tqdm
import pandas as pd
```

### ใส่ค่า rawData , processData และเรียกใช้ datanode

```python
dir_raw_id = 71
dir_process_id = 73
```

```python
datanode = DataNode()
```

### แสดงที่อยู่ข้อมูลและเรียกดูผลข้อมูลที่ได้จากการ Scrapy

```python
meta, fp = datanode.get_file(datanode.get_file_id(directory_id=dir_raw_id, name='turmeric_thestandard.json'))
meta
```

```python
data = json.load(fp)
data
```

### แสดงผลข้อมูลในรูปแบบตาราง

```python
df = pd.json_normalize(data)
df.head()
```

### แสดง Type ของข้อมูล

```python
df[['publish_date', 'retrieve_date']] = df[['publish_date', 'retrieve_date']].astype('datetime64')
df.info()
```

### Upload ลง DataPlatform

```python
datanode.write(df=df, directory=dir_process_id, name="TURMERIC NEWS FROM THE STANDARD", 
               profiling=True, 
               lineage=[datanode.get_file_id(directory_id=dir_raw_id, name='turmeric_thestandard.json')])
```

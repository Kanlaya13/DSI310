### Import function

```python
from dsmlibrary.datanode import DataNode
import json
from tqdm.auto import tqdm
import pandas as pd
```

### ใส่ค่า rawData , processData และเรียกใช้ datanode

```python
dir_raw_id = 71
dir_process_id = 73
```

```python
datanode = DataNode()
```

### แสดงผลข้อมูลและจัดข้อมูลให้อยู่ในรูปแบบตาราง

```python
meta, fp = datanode.get_file(datanode.get_file_id(directory_id=dir_raw_id, name='turmeric.json'))
meta
```

```python
#แสดงข้อมูล
data = json.load(fp)
data
```

```python
#แสดงข้อมูลใรูปแบบตารางเพื่อความสะดวกในการดูข้อมูล
df = pd.DataFrame(data)
df
```

```python
#ได้ทำการ drop ช่องว่างของหัวข้อที่มีการเว้นวรรคกันเนื่องจาก Data Dictionary จะไม่เเสดงผลใน DataPlatform ถ้าไม่ drop ()
#พิกัดศุลกากรมีจำนวน 8 หลัก 
#แก้คำที่มีสระตกหล่นไป
df =df.rename(columns={"พิกัดศุลกากร 8 หลัก" : "พิกัดศุลกากร" })
df =df.rename(columns={"ปริมาณ สถิติ" : "ปริมาณสถิติ" })
df =df.rename(columns={"รหัสสถิต" : "รหัสสถิติ" })
df =df.rename(columns={"หน่วยตามรหัสสถิต" : "หน่วยตามรหัสสถิติ" })
df
```

### Upload datanode

```python
datanode.write(df=df, directory=dir_process_id, name="มูลค่าการส่งออกขมิ้นรายสำนักงานศุลกากรปี2560", profiling=True, lineage=[datanode.get_file_id(directory_id=dir_raw_id, name='turmeric.json')])
```
